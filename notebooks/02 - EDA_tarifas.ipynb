```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from datetime import timedelta

df = pd.read_csv('../data/raw/onibus.csv', delimiter = ';', encoding='unicode_escape')
```

```python
# Converter a coluna 'data' para o formato de data
df['data'] = pd.to_datetime(df['data'], format='%m-%d-%Y')

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df['data'], df['tarifa'], marker='o', linestyle='-', color='b')
plt.title('Gráfico de Tarifas ao longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Tarifa')
plt.grid(True)
plt.show()
```
